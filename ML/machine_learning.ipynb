```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

# Load the dataset
stats = pd.read_csv("player_mvp_stats.csv")

# Add custom accuracy stat
stats['accuracy'] = stats[['FG%', '3P%', '2P%', 'FT%']].mean(axis=1)

# Fill any NaNs with 0 (safe fallback)
stats = stats.fillna(0)

# Strict predictor list (no MVP outcome leaks)
predictors = ['G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
              '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
              'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year', 'W', 'L',
              'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS', 'accuracy']

# Split into train/test
train_df = stats[stats['Year'] < 2024]
test_df = stats[stats['Year'] == 2024].copy()

X_train = train_df[predictors].copy()
X_test = test_df[predictors].copy()
y_train = train_df['Share']

# üí• Custom multipliers (fully respected now)
custom_weights = {
    'PTS': 1.0,
    'AST': 1.5,
    'TRB': 1.2,  # big rebounding multiplier
    'BLK': 1.0,
    'accuracy': 1.0
}

# Apply multipliers (directly affects values)
for stat, mult in custom_weights.items():
    if stat in X_train.columns:
        X_train[stat] *= mult
        X_test[stat] *= mult

# Train Random Forest
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict 2024 MVP shares
test_df['Predicted_Share'] = model.predict(X_test)

# Show Top 5
top_5 = test_df.sort_values(by='Predicted_Share', ascending=False).head(5)
print("\nüèÜ Top 5 Predicted MVP Candidates:")
print(top_5[['Player', 'Tm', 'Predicted_Share']].to_string(index=False))

# Feature importances with custom light blue color
importances = pd.Series(model.feature_importances_, index=X_train.columns)
importances.sort_values(ascending=False).plot(
    kind='bar',
    figsize=(14, 5),
    color='#6BAED6'  # Set the custom light blue color
)
plt.title("Feature Importances in MVP Prediction")
plt.tight_layout()
plt.show()

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Evaluate on training data (or change this to validation data if you want)
y_pred_train = model.predict(X_train)

mae = mean_absolute_error(y_train, y_pred_train)
mse = mean_squared_error(y_train, y_pred_train)
r2 = r2_score(y_train, y_pred_train)

print("\nüìä Error Metrics on Training Data:")
print(f"MAE (Mean Absolute Error): {mae:.4f}")
print(f"MSE (Mean Squared Error): {mse:.4f}")
print(f"R¬≤ Score: {r2:.4f}")
```

```python
# Backtest over range of years
backtest_years = range(2000, 2023)
results = []

for year in backtest_years:
    train_df = stats[stats['Year'] < year]
    test_df = stats[stats['Year'] == year].copy()

    # Skip year if training or testing data is missing
    if train_df.empty or test_df.empty:
        continue

    X_train = train_df[predictors].copy()
    y_train = train_df['Share']
    X_test = test_df[predictors].copy()
    y_test = test_df['Share']

    # Apply custom multipliers
    for stat, mult in custom_weights.items():
        if stat in X_train.columns:
            X_train[stat] *= mult
            X_test[stat] *= mult

    # Train and predict
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Evaluate
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    results.append({'Year': year, 'MAE': mae, 'MSE': mse, 'R2': r2})

# Convert to DataFrame
results_df = pd.DataFrame(results)

# Print all results
print("\nüìà Backtesting Results (2010‚Äì2023):")
print(results_df.to_string(index=False))

# Print averages
avg_mae = results_df['MAE'].mean()
avg_mse = results_df['MSE'].mean()
avg_r2 = results_df['R2'].mean()

print(f"\nüìä Average MAE: {avg_mae:.4f}")
print(f"üìä Average MSE: {avg_mse:.4f}")
print(f"üìä Average R¬≤: {avg_r2:.4f}")
```

```python
import matplotlib.pyplot as plt

# Assuming top_5 is already defined with 'Player' and 'Predicted_Share'
top_5_plot = top_5[['Player', 'Predicted_Share']].copy()

plt.figure(figsize=(10, 6))
plt.barh(top_5_plot['Player'], top_5_plot['Predicted_Share'], color='#6BAED6')
plt.xlabel("Predicted Share")
plt.title("Top 5 Predicted MVPs (2024)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Make sure 'Predicted_Share' exists in test_df (2024)
X_test_2024 = test_df[predictors].copy()
for stat, mult in custom_weights.items():
    if stat in X_test_2024.columns:
        X_test_2024[stat] *= mult

# Predict 2024 if not already done
y_pred_2024 = model.predict(X_test_2024)
test_df['Predicted_Share'] = y_pred_2024

# Get actual vs predicted
y_test_2024 = test_df['Share']

# Error metrics
mae_2024 = mean_absolute_error(y_test_2024, y_pred_2024)
mse_2024 = mean_squared_error(y_test_2024, y_pred_2024)
r2_2024 = r2_score(y_test_2024, y_pred_2024)

# Visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.bar(['MAE', 'MSE', 'R¬≤'], [mae_2024, mse_2024, r2_2024], color='#6BAED6')
plt.title("Error Metrics for 2024")
plt.tight_layout()
plt.show()
```
